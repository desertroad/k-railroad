```kotlin
%useLatestDescriptors
%use dataframe, kandy-geo
```

```kotlin
val stations = DataFrame.readCsv("../data/processed/stations.csv")
    .convert("좌표") {
        "좌표"<DataFrame<*>>()
            .map { "value"<Float>() }
            .takeIf { it.size == 2 }
    }


stations.describe()
```

```kotlin
val skorea = GeoDataFrame.readGeoJson("geo/kr.json")

skorea.plot {
    geoPolygon()

    points {
        stations.filter { 좌표 != null }.run {
            y(map { 좌표!![0] })
            x(map { 좌표!![1] })
        }
    }
}
```
